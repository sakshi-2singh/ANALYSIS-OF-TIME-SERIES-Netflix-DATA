```python
# univariate convlstm example
from numpy import array
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers import ConvLSTM2D
```

```python
# split a univariate sequence into samples
def split_sequence(sequence, n_steps):
    X, y = list(), list()
    for i in range(len(sequence)):
        # find the end of this pattern
        end_ix = i + n_steps
        # check if we are beyond the sequence
        if end_ix > len(sequence)-1:
            break
       # gather input and output parts of the pattern
        seq_x = sequence[i:end_ix]
        seq_y = sequence[end_ix]
        X.append(seq_x)
        y.append(seq_y)
    return array(X), array(y)
```

TO Convert data into list

```python
import csv

def read_column_from_csv(filename, column_index):
    column_data = []
    with open(filename, 'r') as file:
        reader = csv.reader(file)
        for row in reader:
            if len(row) > column_index:
              if row[column_index] == 'Close':
                continue
              else:
                column_data.append(float((row[column_index])))
    return column_data

# Example usage
filename = '/content/drive/MyDrive/archive (10)/NFLX.csv'  # Replace with the actual filename or path
column_index = 4  # Replace with the index of the column you want to extract (zero-based)
column_data = read_column_from_csv(filename, column_index)
#column_data.pop(0)    ##Useless
print(column_data)
```

**Conv_LSTM**

```python
import tensorflow as tf
# define input sequence
raw_seq = column_data
# choose a number of time steps
n_steps = 4
# split into samples
X, y = split_sequence(column_data, n_steps)
# reshape from [samples, timesteps] into [samples, timesteps, rows, columns, features]
n_features = 1
n_seq = 2
n_steps = 2
X = X.reshape((X.shape[0], n_seq, 1, n_steps, n_features))

loss_values = []  # Create an empty list to store loss values
# define model
model = Sequential()
model.add(ConvLSTM2D(filters=64, kernel_size=(1,2), activation='relu', input_shape=(n_seq, 1, n_steps, n_features)))
model.add(Flatten())
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
# fit model
model.fit(X, y, epochs=100, verbose=0, callbacks=[tf.keras.callbacks.LambdaCallback(on_epoch_end=lambda epoch, logs: loss_values.append(logs['loss']))])
# demonstrate prediction

# x_input = array(X[-1])
# x_input = x_input.reshape((1, n_seq, 1, n_steps, n_features))
# yhat = model.predict(x_input, verbose=0)
# print(yhat[0])

y_cap_covlstm=[]
for i in X:
    x_input=np.array(i)
    x_input=x_input.reshape((1, n_seq , 1 ,n_steps,n_features))
    y_cap_covlstm.append(model.predict(x_input,verbose=0))

y_cap_covlstm = [element for sublist in y_cap_covlstm for subsublist in sublist for element in subsublist]
# y_cap_covlstm
last_10_loss = loss_values[-10:]  # Get the last 10 loss values
# Calculate RMSE for each loss value
rmse_values_conlstm = [np.sqrt(loss) for loss in last_10_loss]

print("RMSE values for the last 10 losses:")
print(rmse_values_conlstm)
print(y_cap_covlstm[-1])
```

```python
# univariate cnn lstm example
from numpy import array
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers import TimeDistributed
from keras.layers.convolutional import Conv1D
from keras.layers.convolutional import MaxPooling1D

# split a univariate sequence into samples
def split_sequence(sequence, n_steps):
    X, y = list(), list()
    for i in range(len(sequence)):
        # find the end of this pattern
        end_ix = i + n_steps
        # check if we are beyond the sequence
        if end_ix > len(sequence)-1:
            break
       # gather input and output parts of the pattern
        seq_x = sequence[i:end_ix]
        seq_y = sequence[end_ix]
        X.append(seq_x)
        y.append(seq_y)
    return array(X), array(y)

loss_values = []  # Create an empty list to store loss values
# define input sequence
raw_seq = column_data
# choose a number of time steps
n_steps = 4
# split into samples
X, y = split_sequence(raw_seq, n_steps)
# reshape from [samples, timesteps] into [samples, subsequences, timesteps, features]
n_features = 1
n_seq = 2
n_steps = 2
X = X.reshape((X.shape[0], n_seq, n_steps, n_features))
# define model
model = Sequential()
model.add(TimeDistributed(Conv1D(filters=64, kernel_size=1, activation='relu'), input_shape=(None, n_steps, n_features)))
model.add(TimeDistributed(MaxPooling1D(pool_size=2)))
model.add(TimeDistributed(Flatten()))
model.add(LSTM(50, activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam', loss='mse')
# fit model
model.fit(X, y, epochs=500, verbose=0,callbacks=[tf.keras.callbacks.LambdaCallback(on_epoch_end=lambda epoch, logs: loss_values.append(logs['loss']))])
# demonstrate prediction
# x_input = array(X[-1]))
# x_input = x_input.reshape((1, n_seq, n_steps, n_features))
# yhat = model.predict(x_input, verbose=0)
# print(yhat)

y_cap_cnn=[]
for i in X:
  x_input=np.array(i)
  x_input=x_input.reshape((1,n_seq,n_steps,n_features))
  y_cap_cnn.append(model.predict(x_input,verbose=0))

y_cnn=y.tolist()
y_cap_cnn = [element for sublist in y_cap_cnn for subsublist in sublist for element in subsublist]
last_10_loss = loss_values[-10:]  # Get the last 10 loss values
# Calculate RMSE for each loss value
rmse_values_cnn = [np.sqrt(loss) for loss in last_10_loss]

print("RMSE values for the last 10 losses:")
print(rmse_values_cnn)
print(y_cap_cnn[-1])
```

**Hybrid**

```python
x_hybrid=X
y_hybrid=y
import numpy as np
from keras.models import Model
from keras.layers import Input, LSTM, GRU, Conv1D, MaxPooling1D, Flatten, Dense, concatenate, Bidirectional

n_features = 1
n_steps = x_hybrid.shape[1]

# Reshape the input data
x_hybrid = x_hybrid.reshape((x_hybrid.shape[0], x_hybrid.shape[1], n_features))

# LSTM model
lstm_input = Input(shape=(n_steps, n_features))
lstm = LSTM(50, activation='relu')(lstm_input)

# GRU model
gru_input = Input(shape=(n_steps, n_features))
gru = GRU(50, activation='relu')(gru_input)


# Dense model
dense_input = Input(shape=(n_steps, n_features))
dense = Dense(50, activation='relu')(dense_input)
dense = Flatten()(dense)

# Bidirectional LSTM model
bidirectional_input = Input(shape=(n_steps, n_features))
bidirectional = Bidirectional(LSTM(50, activation='relu'))(bidirectional_input)

# Concatenate the outputs of all models
concatenated = concatenate([lstm, gru, dense, bidirectional])

# Final output layer
output = Dense(1)(concatenated)

# Create the model
model = Model(inputs=[lstm_input, gru_input, dense_input, bidirectional_input], outputs=output)
model.compile(optimizer='adam', loss='mse')

res = []

# Fit the model
history = model.fit([x_hybrid, x_hybrid, x_hybrid, x_hybrid], y, epochs=200, verbose=0)
loss_values = []  # Create an empty list to store loss values

for i in X:
    # Demonstrate prediction
    x_input = np.array(i)
    x_input = x_input.reshape((1, n_steps, n_features))
    y_cap_hybrid = model.predict([x_input, x_input, x_input, x_input], verbose=0)
    res.append(y_cap_hybrid)

y_cap_hybrid = res
y_cap_hybrid = [element for sublist in y_cap_hybrid for subsublist in sublist for element in subsublist]
print(y_cap_hybrid)

last_10_loss = loss_values[-10:]  # Get the last 10 loss values
# Calculate RMSE for each loss value
rmse_values_hybrid= [np.sqrt(loss) for loss in last_10_loss]

print("RMSE values for the last 10 losses:")
print(rmse_values_hybrid)
print(y_cap_hybrid[-1])
```
